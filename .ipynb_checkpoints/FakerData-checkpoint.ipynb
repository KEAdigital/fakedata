```python
# Antal
no_customeres = 20
no_order = 1000
no_employee = 10

# Eksport typer
exp_csv = True           # Gem til CSV
exp_excel = True         # Gem til én Excel fil
exp_sql_script = True    # Gem til SQL Scripts
exp_mysql_azure = True   # Gem direkte til en SQL Server
```

```python
# Connection data
# Hvis der skal eksporteres til en Azure databasse
if exp_mysql_azure:
    from configparser import ConfigParser

    # Læs config file - database.cfg
    parser = ConfigParser()
    _ = parser.read('database.cfg')

    # Data
    username = parser.get('database', 'username')
    password = parser.get('database', 'password')
    host = parser.get('database', 'host')
    port = parser.get('database', 'port')
    db_name = parser.get('database', 'db_name')
```

## Customers

```python
# Customers Dataframe
df_customers = pd.DataFrame(columns=[
    'customer_id',
    'first_name',
    'last_name',
    'address',
    'postcode',
    'city',
    'country',
    'email',
    'date_of_birth'])
```

```python
# Tilføj Customers
for i in range(no_customeres):
    data = [
        i+1,
        fake.first_name(),
        fake.last_name(),  
        fake.address(),
        fake.postcode(),    
        fake.city(),
        fake.country(),
        fake.ascii_safe_email(),
        fake.date_of_birth(minimum_age = 19, maximum_age = 67)]
    df_customers.loc[i] = [item for item in data]
```

## Order

```python
## Order Dataframe
df_order = pd.DataFrame(columns=[
    'order_id',
    'customer_id',
    'orderdate',
    'employee_id'])
```

```python
## Order Dataframe
df_order = pd.DataFrame(columns=[
    'order_id',
    'customer_id',
    'orderdate',
    'employee_id'])
```

```python
# Tilføj Order
for i in range(no_order):
    data = [
        i+1,
        fake.random_int(0, 200),    
        fake.date_between(start_date='-3y', end_date='now'),
        fake.random_int(1, no_employee)]
    df_order.loc[i] = [item for item in data]
```

```python
# Tilføj Deliverydate
df_order['deliverydate'] = df_order['orderdate'] + pd.DateOffset(days=fake.random_int(0, 12))
```

```python
#df_order.head()
df_order.tail()
```

## Order-Line

```python
## OrderLine Dataframe
df_orderline = pd.DataFrame(columns=[
    'order_id',
    'product_id',
    'quantity'])
```

```python
# Tilføj Orderline
for i in range(no_order + 10000):
    data = [
        fake.random_int(10000, 15000),
        fake.random_int(1, 20),
        fake.random_int(1, 100)]
    df_orderline.loc[i] = [item for item in data]
```

# Export
Export af data til CSV, Excewl og SQL.

I Excel gemmes der én Excel fil, med data fordelt på 5 ark.

I CSV gemmes der til en CSV filer pr. dataframe - altså 5 csv filer.

```python
# Gem til CSV
if exp_csv:
    df_customers.to_csv('customers.csv', sep=';', index=False)
    df_order.to_csv('order.csv', sep=';', index=False, decimal=',')
    df_orderline.to_csv('orderline.csv', sep=';', index=False, decimal=',')
    df_employee.to_csv('employees.csv', sep=';', index=False)
    df_products.to_csv('products.csv', sep=';', index=False)
```

```python
# Gem til Excel
if exp_excel:
    create_excel = pd.ExcelWriter('sales_data.xlsx', engine='xlsxwriter', date_format='dd-mm-yyyy')

    # Overfør hver DF til Excel
    df_customers.to_excel(create_excel, sheet_name='customers', index=False)
    df_order.to_excel(create_excel, sheet_name='order', index=False)
    df_orderline.to_excel(create_excel, sheet_name='orderline', index=False)
    df_employee.to_excel(create_excel, sheet_name='employee', index=False)
    df_products.to_excel(create_excel, sheet_name='products', index=False)

    # Gem Excel fil
    create_excel.save()
```

# sqlalchemy
SQLAlchemy is the Python SQL toolkit and Object Relational Mapper that gives application developers the full power and flexibility of SQL.

https://www.sqlalchemy.org/

```python
if exp_mysql_azure:
    # Opret forbindelse
    from sqlalchemy import create_engine
    from sqlalchemy_utils import database_exists, create_database
```

```python
# Create database hvis den ikke findes
if exp_mysql_azure:
    # Connection
    my_conn = create_engine(f"mysql://{username}:{password}@{host}:{port}")

    with my_conn.connect() as conn:
        conn.execute(f"CREATE DATABASE IF NOT EXISTS {db_name}")

    # Vælg database
    my_conn.execute(f"USE {db_name}")
```

```python
if exp_mysql_azure:
    # Opret tabeller og overfør data
    # Der er en bug i Pandas der gør at if_exists="replace" ikke virker!!
    # nemmest løsning er at slette databasen (mySalesData) via Workbench
    
    # Products
    df_products.to_sql('tblProducts', my_conn, if_exists='append', index=False)

    # Customers
    df_customers.to_sql('tblCustomers', my_conn, if_exists='append', index=False)

    # Order
    df_order.to_sql('tblOrder', my_conn, if_exists='append', index=False)

    # Orderline
    df_orderline.to_sql('tblOrderlines', my_conn, if_exists='append', index=False)

    # Employee
    df_employee.to_sql('tblEmployee', my_conn, if_exists='append', index=False)
```

```python
 # Primary Key
if exp_mysql_azure:
    with my_conn.connect() as conn:
        my_conn.execute('USE mySalesData;')
        my_conn.execute('ALTER TABLE tblProducts ADD CONSTRAINT PK_Products PRIMARY KEY (product_id);')
        my_conn.execute('ALTER TABLE tblCustomers ADD CONSTRAINT PK_customers PRIMARY KEY (customer_id);')
        my_conn.execute('ALTER TABLE tblOrder ADD CONSTRAINT PK_order PRIMARY KEY (order_id);')
        # my_conn.execute('ALTER TABLE tblOrderlines ADD CONSTRAINT PK_orderlines PRIMARY KEY (product_id);')
        my_conn.execute('ALTER TABLE tblEmployee ADD CONSTRAINT PK_employee PRIMARY KEY (employee_id);')
```

```python
# Chack om der er overført data
pd.read_sql("select * from products;", my_conn)
```
